**Feature Engineering & KPI Creation**

```python
# ------------------------------
# Total Revenue
# ------------------------------
dataset['TotalRevenue'] = dataset['TotalAmount']


# ------------------------------
# Average Order Value (AOV)
# ------------------------------
# Average Order Value = Total Revenue Ã· Number of Unique Orders
aov = dataset.groupby('OrderID')['Revenue'].sum().mean()
print("Average Order Value (AOV):", round(aov, 2))


# ------------------------------
# Customer Lifetime Value (CLV):
# CLV=AOVÃ—PurchaseFrequencyÃ—RetentionPeriod
# ------------------------------

# Purchase Frequency = Total Orders Ã· Total Customers
total_orders = dataset['OrderID'].nunique()
total_customers = dataset['CustomerID'].nunique()
purchase_frequency = total_orders / total_customers

# Retention Period (e.g., assume 1 year for simplicity)
retention_period = 1

# CLV = AOV Ã— Purchase Frequency Ã— Retention Period
clv = aov * purchase_frequency * retention_period
print("Estimated Customer Lifetime Value (CLV):", round(clv, 2))

# ------------------------------
# Gross Revenue Growth (Monthly)
# ------------------------------

# Ensure OrderDate is datetime with the correct format
dataset['OrderDate'] = pd.to_datetime(dataset['OrderDate'], format='%d/%m/%Y')

# Extract Year-Month
dataset['YearMonth'] = dataset['OrderDate'].dt.to_period('M')

# Monthly Revenue
monthly_revenue = dataset.groupby('YearMonth')['TotalAmount'].sum().reset_index()

# Calculate Revenue Growth %
monthly_revenue['RevenueGrowth(%)'] = monthly_revenue['TotalAmount'].pct_change() * 100

print("\nMonthly Revenue Growth:")
print(monthly_revenue)
```

```python
# ------------------------------
# Feature Importance Analysis
# ------------------------------
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import plotly.express as px

# 1. Prepare the dataset
df = dataset.copy()

# Encode categorical features
for col in df.select_dtypes(include='object').columns:
    if col != 'OrderID':  # Exclude IDs
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col].astype(str))

# Define target and features
target = 'TotalAmount'
features = [col for col in df.columns if col != target]

X = df[features]
y = df[target]

# 2. Train Random Forest Regressor
rf = RandomForestRegressor(n_estimators=200, random_state=42)
rf.fit(X, y)

# 3. Get feature importances
importances = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf.feature_importances_
}).sort_values(by='Importance', ascending=False)

# 4. Plot feature importance (interactive)
fig = px.bar(importances,
             x='Importance',
             y='Feature',
             orientation='h',
             color='Importance',
             color_continuous_scale='Viridis',
             text='Importance',
             title='Feature Importance for Total Sales Amount (Random Forest)')
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```
